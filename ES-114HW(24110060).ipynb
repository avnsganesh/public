```python
import math
list1=[]
for i in range(1,21):
    list1+=[i]
sue=0
for j in list1[1::2]:
    sue+=j
print(sue)
suo=1
for k in list1[::2]:
    suo*=k
print(suo)
print(sum(list1)/20)
median=(list1[(len(list1)//2)-1]+list1[len(list1)//2])/2
print(median)
sus=0
for l in list1:
    sus+=(l-median)**2
print(math.sqrt(sus/20))
```

```python
def fibonacci(x):
    list1=[0]
    if x==1:
        return list1
    list1=[0,1]
    if x==2:
        return list1
    for i in range(3,x+1):
        list1+=[fibonacci(i-1)[i-3]+fibonacci(i-1)[i-2]]
    return list1
n=int(input())
print(fibonacci(n))
from matplotlib import pyplot as plt
a=[i for i in range(1,n+1)]
b=fibonacci(n)
plt.plot(a,b,marker='o')
plt.title('fibonacci sequence line graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks(a)
plt.yticks(b)
plt.grid()
plt.show()
```

```python
def string(x):
    lower_string=x.lower()
    dicts={}
    for k in lower_string:
        dicts[k]=0
    for i in lower_string:
        dicts[i]+=1
    return dicts
n=input()
print(string(n))
from matplotlib import pyplot as plt
plt.bar([i for i in string(n)],[string(n)[i] for i in string(n)])
plt.yticks([string(n)[i] for i in string(n)])
plt.show()
```

```python
def strings(x):
    longest_string=''
    ma=0
    s=0
    listo=['a','e','i','o','u']
    for i in x:
        if len(i)>=ma:
            ma=len(i)
            longest_string=i
        if i[0].lower() in listo:
            s+=1
    print(f'longest string :{longest_string} , no of strings started from vowel :{s}')
    return ma    
list1=['asdf','qwergh','erfv','okjhgvc','mn','jklkjhgfds','iuhgv']
ma=strings(list1)
from matplotlib import pyplot as plt
plt.hist([len(i) for i in list1],bins=[i for i in range(0,ma+1)])
plt.xticks([i for i in range(0,ma+1)])
plt.show()
```

```python
def sort_dict(x):
    list1=[]
    for i in x:
        list1+=[i['age']]
    new_x=[]
    for j in sorted(list1):
        for l in x:
            if l['age']==j:
                new_x+=[l]
                break
    print(new_x)
data=[{'name': 'alice','age':25},{'name':'bob','age':22}]
sort_dict(data)
```
